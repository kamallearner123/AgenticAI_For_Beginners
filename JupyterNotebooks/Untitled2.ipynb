```python
import numpy as np
# Model: Linear regression
from sklearn.linear_model import LinearRegression


# Input
hum = np.array([70, 60, 40, 50, 90]).reshape(-1, 1)
# Actual output
rain = np.array([0.8, 0.7, 0.4, 0.5, 0.8])


# Get the model
model = LinearRegression()
model.fit(hum, rain)# Training

print(model.predict([[55],[45]]))
print(model.predict([[30]]))
print(model.coef_) # weights y = x*0.014

import matplotlib.pyplot as plot
plot.scatter(hum, rain)
guess = [[10],[55],[45], [90]]
plot.plot(guess, model.predict(guess), color = 'red')
plot.xlabel("Humidity")
plot.ylabel("Rain fall")
plot.title("Rain fall prediction")
```

```python
import numpy as np
# Model: Linear regression
from sklearn.linear_model import LinearRegression


# Input
hum = np.array([[30, 70], [30, 60], [30, 40], [30, 50], [30, 90]])
# Actual output
rain = np.array([0.8, 0.7, 0.4, 0.5, 0.8])


# Get the model
model = LinearRegression()
model.fit(hum, rain)# Training

print(model.predict([[30, 55],[40, 45]]))
print(f" Coefficient of Temp = {model.coef_[0]}, Coefficient of Humidity = {model.coef_[1]}") 
# Note: First parameter does not show any impact on output
```
